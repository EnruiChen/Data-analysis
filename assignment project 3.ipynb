### Question 1

First we import the csv files of Cobalt (Co) and Rhodium (Rh) using command 'read_csv()' 

```python
import pandas as pd
co_molecule_df = pd.read_csv("data/Co_CN4-molecules.csv", 
                             sep = ",",
                            na_values = [-999], # the missing data are shown as '-999', we changed it into 'NaN'
                            skiprows = 1,
                            index_col = ["Refcode"])
co_molecule_df.head()
```

```python
rh_molecule_df = pd.read_csv("data/Rh_CN4-molecules.csv", 
                             sep = ",",
                            na_values = [-999], # the missing data are shown as '-999', we changed it into 'NaN'
                            skiprows = 1,
                            index_col = ["Refcode"])
rh_molecule_df.head()
```

### Question 2

Now we caculate the mean angle of each molecules, and put the value into a columns called "Average Angle"

```python
# Calculate the average angle in each row.
co_molecule_df["Average Angle"] = (co_molecule_df["ANG1"] + co_molecule_df["ANG2"] 
                                   + co_molecule_df["ANG3"] + co_molecule_df["ANG4"])/4 
co_molecule_df.head()
```

```python
# Calculate the average angle in each row.
rh_molecule_df["Average Angle"] = (rh_molecule_df["ANG1"] + rh_molecule_df["ANG2"] 
                                   + rh_molecule_df["ANG3"] + rh_molecule_df["ANG4"])/4
rh_molecule_df.head()
```

Calculate the standard deviation for the average angle. And compare to Square Planar ($ 90^{\text{o}} $) and Tetrahedral shapes ($ 109.5^{\text{o}}$).

```python
co_average_angle = co_molecule_df["Average Angle"] 
co_angle_std = co_average_angle.std() # Calculate the standard deviation of molecule of Co
co_angle_mean = co_average_angle.mean() # Calculate the mean value of molecule of Co

print(f"The standard deviation of Cobalt (Co) element is {co_angle_std:.3f}, the mean value of angle is {co_angle_mean:.3f}")

co_difference_1 = 109.5 - co_angle_mean
co_difference_2 = co_angle_mean - 90
if co_angle_mean > 109.5:
    print(f"The difference between Tetrahedral shapes (109.5) is {co_difference_1:.3}, and between Square Planar is {co_difference_2:.3} ")
else:
    print(f"Since the mean value of angle is smaller than 109.5, so the difference between mean angle and the Square Planner is {co_difference_2:.3}") 
```

```python
rh_average_angle = rh_molecule_df["Average Angle"]
rh_angle_std = rh_average_angle.std() # Calculate the standard deviation of angle of molecule Rh
rh_angle_mean = rh_average_angle.mean() # Caculate the mean value of anlge of molecule Rh

print(f"The standard deviation of Rhodium (Rh) element is {rh_angle_std:.3f}, the mean value of angle is {rh_angle_mean:.3f}")

rh_difference_1 = 109.5 - rh_angle_mean
rh_difference_2 = rh_angle_mean - 90
if co_angle_mean > 109.5:
    print(f"The difference between Tetrahedral shapes (109.5) is {rh_difference_1:.3}, and between Square Planar is {rh_difference_2:.3} ")
else:
    print(f"Since the mean value of angle is smaller than 109.5, so the difference between mean angle and the Square Planner is {rh_difference_2:.3}") 
```

### Question 3

```python
# Create lists of name of "Refcode" values: 
molecules_list_rh = ['YIQQUK']
molecules_list_co = ['ZOJPUJ01', 'ICAYES', 'ZABVIK']

# Using the list that creat above and using loop to find the value of it. 
for refcode_rh in molecules_list_rh:
    filt_rh = rh_molecule_df.index == refcode_rh # Create a filter.
    rh_select_molecule_df = rh_molecule_df[filt_rh]

co_select_molecule_df = [] # Create a dataframe. 
for refcode_co in molecules_list_co:
    filt_co =(co_molecule_df.index == refcode_co)   # Create a filter
    co_select_molecule_df.append(co_molecule_df[filt_co]) # Apply filter and append to the dataframe create before

selected_molecules_df = rh_select_molecule_df.append(co_select_molecule_df)
selected_molecules_df
```

```python
average_angle = selected_molecules_df["Average Angle"] # Extract "Average Angle" column
index = average_angle.index # Create list of Refcode that's index
n = 0

for angle in average_angle:
    difference_1 = abs(90 - angle) # Differnce between bond angle and Square Planar
    difference_2 = abs(109.5 - angle) # Difference between bond angle and Tetrahedral shape
    if difference_1 < difference_2:
        print(
            f"The molecule of Refcode {index[n]}, average angle: {angle:.2f} is more likely to having a Square planar")
    else:
        print(
            f"The molecule of Refcode {index[n]}, average angle: {angle:.2f} is more likely to having a Tetrahedral shape")
    n = n + 1
```
